### Oppgave 1a)
For å vise at
$$
\phi(x,t)=\frac{1}{\sqrt{4\pi Dt}}\int_{-\infty}^\infty \mathrm{dy} \, h(y) \, e^{-\frac{(x-y)^2}{4Dt}}
$$

løser
$$
\frac{∂\phi}{∂t}=D\frac{∂^2\phi}{∂x^2} \hspace{15mm} \text{(1)}
$$

defineres først $C(t)=\frac{1}{\sqrt{4\pi Dt}}$ og $f(x, y, t)=h(y) \, e^{-\frac{(x-y)^2}{4Dt}}$, slik at $\phi=C*\int_{-\infty}^\infty f\mathrm{dy}$. De aktuelle partiellderiverte er vist under, der vi har brukt kjerneregelen og produktregelen for derivasjon.

$\frac{∂C}{∂t}=-\frac{1}{\sqrt{4\pi D}}*\frac{1}{2t^{3/2}}=-\frac{C}{2t}$

$\frac{∂f}{∂t}=\frac{(x-y)^2}{4Dt^2}f$

$\frac{∂f}{∂x}=-\frac{2(x-y)}{4Dt}f=-\frac{(x-y)}{2Dt}f$

$\frac{∂^2f}{∂x^2}=-\frac{1}{2Dt}f-\frac{(x-y)}{2Dt}*\frac{∂f}{∂x}=-\frac{1}{2Dt}f+\frac{(x-y)}{2Dt}*\frac{(x-y)}{2Dt}f=f\left(\frac{(x-y)^2}{4D^2t^2}-\frac{1}{2Dt}\right)$

Anta at $h(y)$ er slik at $\int_{-\infty}^\infty f\mathrm{dy}$ er leddvis deriverbar. Da er $\frac{∂\phi}{∂t}=\frac{∂}{∂t}\left(C*\int_{-\infty}^\infty f\mathrm{dy}\right)=\frac{∂C}{∂t}*\int_{-\infty}^\infty f\mathrm{dy}+C*\frac{∂}{∂t}\int_{-\infty}^\infty f\mathrm{dy}=\frac{∂C}{∂t}*\int_{-\infty}^\infty f\mathrm{dy}+C*\int_{-\infty}^\infty \frac{∂f}{∂t}\mathrm{dy}$

$=-\frac{C}{2t}\int_{-\infty}^\infty f\mathrm{dy}+C\int_{-\infty}^\infty \frac{(x-y)^2}{4Dt^2}f\mathrm{dy}=\underline{C\int_{-\infty}^\infty f\left(\frac{(x-y)^2}{4Dt^2}-\frac{1}{2t}\right)\mathrm{dy}}$

$\frac{∂^2\phi}{∂x^2}=C*\frac{∂^2}{∂x^2}\int_{-\infty}^\infty f\mathrm{dy}=C\int_{-\infty}^\infty \frac{∂^2f}{∂x^2}\mathrm{dy}=\underline{C\int_{-\infty}^\infty f\left(\frac{(x-y)^2}{4D^2t^2}-\frac{1}{2Dt}\right)\mathrm{dy}=\frac{1}{D}*\frac{∂\phi}{∂t}}$.

$\phi$ løser dermed (1).

Løsninga er en konvolusjon av funksjonene $h(x)$ og $\frac{1}{\sqrt{4\pi Dt}}e^{-\frac{x^2}{4Dt}}$. Sistnevnte er normalfordelinga sentrert i $x=0$ og med tidsavhengig varians $σ^2=2Dt$. Ved å la $t→0$ går normalfordelinga mot en delta-funksjon, og vi ser at $h(x)$ er initialtilstanden til systemet:

$\phi(x,0)=\int_{-\infty}^\infty h(y)*\lim_{t\to0}\frac{1}{\sqrt{4\pi Dt}}e^{-\frac{(x-y)^2}{4Dt}}\mathrm{dy}=\int_{-\infty}^\infty h(y)*δ(x-y)\mathrm{dy}=h(x).$

Her har vi altså brukt relasjonen $\lim_{σ\to0}𝒩(μ,σ)=δ(x-μ)$. Løsninga består dermed av den opprinnelige fordelinga h(x) og en normalfordeling som "smøres" utover langs x ettersom variansen øker med tid.

### Oppgave 1b)

Sannsynlighetene for p-, p og p+ er gitt av sannsynlighetene for å hoppe et steg bak, stå i ro og et steg fram, som vi antar er proporsjonal med P(x).

Antar at sannsynligheten for å ta et steg fra $x_0$ til $x$ er proporsjonal med $P(x)∝e^{-βV(x)}$, med samme proporsjonalitetskonstant for alle mulige steg. Da er

$p^- \propto P(x_0-1)∝e^{-βV(x_0-1)} ⟹ p^-=C*e^{-βV(x_0-1)}$

$p^0 \propto P(x_0)∝e^{-βV(x_0)} \hspace{13mm} ⟹ p^0=C*e^{-βV(x_0)}$

$p^+ \propto P(x_0+1)∝e^{-βV(x_0+1)} ⟹ p^+=C*e^{-βV(x_0+1)}$.

Total sannsynlighet $1=p^-+p^0+p^+=C\left(e^{-βV(x_0-1)}+e^{-βV(x_0)}+e^{-βV(x_0+1)}\right) ⟹ C=\left(e^{-βV(x_0-1)}+e^{-βV(x_0)}+e^{-βV(x_0+1)}\right)^{-1}$. Innsetting av $C$ i uttrykkene over gir

$p^-=\left[\left(e^{-βV(x_0-1)}+e^{-βV(x_0)}+e^{-βV(x_0+1)}\right)e^{βV(x_0-1)}\right]^{-1}=\left(1+e^{-β[V(x_0)-V(x_0-1)]}+e^{-β[V(x_0+1)-V(x_0-1]}\right)^{-1}$

$p^0=\left[\left(e^{-βV(x_0-1)}+e^{-βV(x_0)}+e^{-βV(x_0+1)}\right)e^{βV(x_0)}\right]^{-1}=\left(e^{-β[V(x_0-1)-V(x_0)]}+1+e^{-β[V(x_0+1)-V(x_0)]}\right)^{-1}$

$p^+=\left[\left(e^{-βV(x_0-1)}+e^{-βV(x_0)}+e^{-βV(x_0+1)}\right)e^{βV(x_0+1)}\right]^{-1}=\left(e^{-β[V(x_0-1)-V(x_0+1)]}+e^{-β[V(x_0)-V(x_0+1)]}+1\right)^{-1}$.

### Oppgave 1c)

$k_B T \gg |V(x+1)-V(x)| \Rightarrow β|V(x+1)-V(x)|\ll1$. Følgelig er $e^{\pmβ[V(x+1)-V(x)]}≈1$. Det samme gjelder

$e^{\pmβ[V(x+1)-V(x-1)]}=e^{\pmβ[V(x+1)-V(x)]}*e^{\pmβ[V(x)-V(x-1)]}$. Innsatt i uttrykkene for sannsynlighetene {$p^+$, $p^0$, $p^-$} gir dette

$
p^{+} = \frac{1}{3} \\
p^{0} = \frac{1}{3} \\
p^{-} = \frac{1}{3}
$

Her er altså systemets termiske energi så høy at alle energitilstander er like sannsynlige.

La oss nå i stedet se på tilfellet der $k_B T \ll |V(x+1)-V(x)|$ for alle $x$. Vi definerer $[V(x+1)-V(x)] = δ $. Siden $δ\gg k_BT$, er variasjoner i $δ$ irrelevante og vi setter den konstant for alle $x$. Da blir også nødvendigvis $[V(x)-V(x+1)]= -\delta$, og $[V(x+1)-V(x-1)] = 2\delta$.
Vi setter dette inn i uttrykkene våre for $p^-$, $p^0$ og $p^+$ og ser hva vi får.



$p^-=\left(1+e^{-β[V(x_0)-V(x_0-1)]}+e^{-β[V(x_0+1)-V(x_0-1]}\right)^{-1}$
$=\left(1+e^{-\beta\delta}+e^{-2\beta\delta}\right)^{-1}$

$p^0=\left(e^{-β[V(x_0-1)-V(x_0)]}+1+e^{-β[V(x_0+1)-V(x_0)]}\right)^{-1}$
$=\left(e^{\beta\delta}+1+e^{-\beta\delta}\right)^{-1}$

$p^+=\left(e^{-β[V(x_0-1)-V(x_0+1)]}+e^{-β[V(x_0)-V(x_0+1)]}+1\right)^{-1}$
$=\left(e^{2\beta\delta}+e^{\beta\delta}+1\right)^{-1}$

Den aktuelle grensa er nå at $T\to0⇔β\to∞$. Om vi lar $V(x)$ øke med $x$ slik at $\delta > 0$ får vi at eksponenisalfunksjonene på  formen $e^{-βδ}$ går mot $0$ og alle på formen $e^{βδ}$ går mot $∞$. Da følger det at

$p^- \to 1$,
$p^0 \to 0$ og
$p^+ \to 0$.

Om vi derimot skulle ha tenkt at $V(x)$ sank med økende $x$ blir $\delta < 0$ og med samme argumentasjon som forrige tilfelle ville vi fått at

$p^- \to 0$,
$p^0 \to 0$ og
$p^+ \to 1$.

Begge tilfeller kan vi tolke likt. Den karakteristiske termiske energien til systemet, $k_BT$, setter begrensninger for energien til partiklene i systemet. Dersom denne er mye lavere enn endringen i potensial over en skrittlengde, er det usannsynlig at partiklene beveger seg mot, og forblir i, høyere energitilstander.

I de to ovennevnte grensene trenger vi altså ikke å beregne sannsynligheten for hvert steg. I høytemperaturgrensa er sannsynlighetsfordelinga med god tilnærming en uniformfordeling, og i lavtemperaturgrensa dominerer sannsynligheten for å gå mot lavere potensial.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random as rd
import tqdm

plt.rcParams.update({
    'figure.dpi': 110,
    'font.family': 'serif',
    'mathtext.fontset': 'cm',
    'axes.grid': True,
    'lines.linewidth': 0.8,
})
```

```python
def rw1(positions, bk, amount_places, amount_particles, iterations):
    # Constant potential
    p_plus = 1/3
    p_minus = 1/3
    positions_new = positions.copy()
    for r in range(iterations):
        for i in range(amount_places):
            if positions[i] == 0: continue
            for j in range(int(positions[i])):
                q = rd.uniform(0,1)
                if q <= p_minus:
                  if i > 0:
                    positions_new[i-1] += 1
                    positions_new[i] -= 1
                elif q > (1-p_plus):
                  if i < (amount_places - 1):
                    positions_new[i+1] += 1
                    positions_new[i] -= 1
        positions = positions_new.copy()
    return positions


def rw2(positions, bk, amount_places, amount_particles, iterations):
    # W. potential gradient
    positions_new = positions.copy()
    for r in range(iterations):
        for i in range(amount_places):
            if positions[i] == 0: continue
            for j in range(int(positions[i])):
                p_minus = (1 + np.e**(bk*2) + np.e**(bk))**(-1)
                p_plus = (1 + np.e**(-bk*2)+np.e**(-bk))**(-1)
                q = rd.uniform(0,1)
                if q <= p_minus:
                  if i > 0:
                    positions_new[i-1] += 1
                    positions_new[i] -= 1
                elif q > (1-p_plus):
                  if i < (amount_places - 1):
                    positions_new[i+1] += 1
                    positions_new[i] -= 1
        positions = positions_new.copy()
    return positions

def main(version, bk = 1, amount_places = 1001, amount_particles = 101, iterations = 200):

    positions = np.zeros(amount_places)
    start_pos_place_particles = int(math.floor(amount_places/2) - math.floor(amount_particles/2))
    for i in range(amount_particles):
        positions[start_pos_place_particles+i] = 1       # Particles placed at the centre of the system

    if version == "v1":
        positions = rw1(positions, bk, amount_places, amount_particles, iterations)
    elif version == "v2":
        positions = rw2(positions, bk, amount_places, amount_particles, iterations)

    return positions

def avg_pos(version, bk = 1, amount_places = 1001, amount_particles = 101, iterations = 200, epochs = 100):
  # Epochs: Amount of calls to the random walk function
  positions = main(version, bk, amount_places, amount_particles, iterations)

  for i in range(epochs-1):
    positions += main(version, bk, amount_places, amount_particles, iterations)

  return positions/epochs
```

```python
total_steps = 200

v1_avg = avg_pos('v1', iterations = total_steps)
v2_avg = avg_pos('v2', iterations = total_steps)
```

```python
# Resultat
N = len(v1_avg)

plt.plot(v1_avg, label ='$V(x)=k$')
plt.plot(v2_avg, label ='$V(x)=-kx$')
plt.xticks([0, N/4, N/2,(3*N)/4,N], [f'{-N/2:.0f}', f'{-N/4:.0f}', '0', f'{N/4:.0f}', f'{N/2:.0f}'])
plt.legend()
plt.title(f'Partikkelfordeling etter t={total_steps}')
# Limit y to the smallest peak, in case one is way larger than the other
plt.ylim(bottom=0, top = np.min([np.max(v1_avg), np.max(v1_avg)])+0.1)
plt.show()
```

### Oppgave 2b)

```python
class ParticleMoveFunctions:
    """
    Functions used to move particles with different potential functions.
    """
    def k(self):
        p_minus = 1/3
        p_plus = 1/3
        q = random.uniform(0,1)
        if q <= p_minus: return "left"
        elif q > 1-p_plus: return "none"
        else: return "right"

    def minus_k_x(self) -> str:
        p_minus = (1 + np.e**(2) + np.e)**(-1)
        p_plus = (1 + np.e**(-2)+np.e**(-1))**(-1)
        q = random.uniform(0,1)
        if q <= p_minus: return "left"
        elif q > (1 - p_plus): return "right"
        else: return "none"

    def very_strong_potential(self) -> str:
        q = random.uniform(0,1)
        return "right"
        if q <= 0.0: return "left"
        elif q > (0): return "right"
        else: return "none"
```

```python
def place_particles(start_pos: int = 0, amount_of_particles = 101, line_steps = 1001) -> np.ndarray:
    """
    Place particles on a line.
    """
    line = np.zeros(line_steps)
    particles = amount_of_particles

    if line_steps - start_pos < particles:
        raise ValueError

    for i in range(amount_of_particles):
        line[start_pos+i] = 1

    return line
```

```python
def random_walk_block(
        func,
        num_steps: int = 200,
        particle_line: np.ndarray = place_particles(),
        ) -> np.ndarray:
    """
    Simulate each particle walking randomly, but particles can't occupy same position.

    Parameters:
        func: Function that returns a direction ("left", "right", "none").
    """

    get_step_direction = func
    number_of_particles = sum(particle_line)

    particle_line_new = particle_line.copy()
    for i in range(num_steps):
        # List of the index of each particle
        random_particle_index_list = [index for index, item in enumerate(particle_line) if item == 1]
        random.shuffle(random_particle_index_list)

        for j in random_particle_index_list:

            direction = get_step_direction()

            if direction == "left":
                if not j-1 >= 0: continue
                if particle_line_new[j-1] == 1: continue
                particle_line_new[j] = 0
                particle_line_new[j-1] = 1

            elif direction == "right":
                if not j+1 < len(particle_line): continue
                if particle_line_new[j+1] == 1: continue
                particle_line_new[j] = 0
                particle_line_new[j+1] = 1

        particle_line = particle_line_new.copy()

    return particle_line
```

```python
epochs = 200
len_line = 1001
number_of_particles = 100

# get the move function
k = ParticleMoveFunctions().k
minus_k_x = ParticleMoveFunctions().minus_k_x

# potential V(x) = k
positions = np.zeros(len_line)
for i in range(epochs-1):
    positions += random_walk_block(k, particle_line=place_particles(start_pos=450))
mean = (positions / epochs)

# potential V(x) = -kx
positions2 = np.zeros(len_line)
for i in range(epochs-1):
    positions2 += random_walk_block(minus_k_x, particle_line=place_particles(start_pos=450))
mean2 = (positions2 / epochs)


# 2a) plots
plt.plot(v1_avg, label ='$V(x)=k$')
plt.plot(v2_avg, label ='$V(x)=-kx$')

# 2b) plots
plt.plot(mean, label="$V(x) = k$ (interaksjon)", color = 'k')
plt.plot(mean2, label="$V(x) = -kx$ (interaksjon)")
plt.title(f'Partikkelfordeling etter t={epochs}.')
plt.xlabel("Posisjon")
plt.ylabel("Gjennomsnittlig partikkelantall")

N = len(v1_avg)
plt.xticks([0, N/4, N/2,(3*N)/4,N], [f'{-N/2:.0f}', f'{-N/4:.0f}', '0', f'{N/4:.0f}', f'{N/2:.0f}'])

# Limit y to the smallest peak, in case one is way larger than the other
plt.ylim(bottom=0, top = np.min([np.max(v1_avg), np.max(v1_avg)])+0.1)

plt.legend()
plt.show()
```

Det er tydelig at partiklene flytter seg mindre når de ikke kan dele plass. For konstant potensial (svart graf) likner partikkelfordelinga svært mye på initialfordelinga, altså har ikke partiklene flytta seg mye i det hele tatt. For synkende potensial langs x (grønn graf) er fordelinga forskjøvet mot høyre som forventa, men i mindre grad enn når partiklene ikke interagerer. Begge disse resultatene er forenelige med det at partiklene ikke kan dele plass. Resultatet av dette er at det til å begynne med kun er partiklene på hver ende av fordelinga som kan flytte seg, imens mesteparten av dem står stille i midten og "venter".

```python
def ratchet(x, alpha, N):
  # Alpha determines the shape of the sawtooth pattern
  # N: The spacial period of the potential

  if 0 < x <= alpha*N:
    V_1 = x/(alpha*N)

  elif -(1-alpha)*N <= x <= 0:
    V_1 = -x/((1-alpha)*N)

  else:
    print('Input outside of function domain')

  return V_1

def periodic(func, start, period, x, *params):
  # Turns a function (func) defined on the interval
  # [start, start+period] into a periodic function

  # params (optional): Tuple containing the parameters for func, excluding x

  n = len(x)
  out = np.zeros(n)

  for i in range(n):
    left_of_interval = (x[i] < start)
    right_of_interval = (x[i] > (start+period))

    if left_of_interval:
      x_mod = x[i] + period
      while x_mod < start:
        x_mod += period
      out[i] = func(x_mod, *params)

    elif right_of_interval:
      x_mod = x[i] - period
      while x_mod > (start+period):
        x_mod -= period
      out[i] = func(x_mod, *params)

    else:
      out[i] = func(x[i], *params)

  return out
```

```python
def periodic_step(positions, V, bk, amount_places, amount_particles):
    # W. potential gradient
    n_plus = 0
    n_minus = 0
    positions_new = positions.copy()
    for i in range(amount_places):
        if positions[i] == 0: continue
        for j in range(int(positions[i])):
            if 0 < i < (amount_places-1): #checks if between boundaries
                p_minus = (1 + np.e**(-bk*(V[i+1]-V[i-1])) + np.e**(-bk*(V[i]-V[i-1])))**(-1)
                p_plus = (1 + np.e**(-bk*(V[i-1]-V[i+1])) + np.e**(-bk*(V[i]-V[i+1])))**(-1)

            elif i == 0: #checks if at left boundary
                p_minus = (1 + np.e**(-bk*(V[i+1]-V[amount_places-1])) + np.e**(-bk*(V[i]-V[amount_places-1])))**(-1)
                p_plus = (1 + np.e**(-bk*(V[amount_places-1]-V[i+1])) + np.e**(-bk*(V[i]-V[i+1])))**(-1)

            elif i == (amount_places-1): #checks if at right boundary
                p_minus = (1 + np.e**(-bk*(V[0]-V[i-1])) + np.e**(-bk*(V[i]-V[i-1])))**(-1)
                p_plus = (1 + np.e**(-bk*(V[i-1]-V[0]))+np.e**(-bk*(V[i]-V[0])))**(-1)

            q = rd.uniform(0,1)
            if q <= p_minus:
                if i > 0:
                    positions_new[i-1] += 1
                    positions_new[i] -= 1
                else:
                    positions_new[amount_places-1] += 1
                    positions_new[i] -= 1
                n_minus += 1
            elif q > (1-p_plus):
                if i < (amount_places - 1):
                    positions_new[i+1] += 1
                    positions_new[i] -= 1
                else:
                    positions_new[0] += 1
                    positions_new[i] -= 1
                n_plus += 1
    J = (n_plus - n_minus)/amount_particles
    positions = positions_new.copy()
    return positions, J


def const_step(positions, amount_places, amount_particles):
    """
    One step with constant potential $V(x) = k$

    Params:
        positions: list with amount of particles at each step.
    """
    n_minus = 0
    n_plus = 0
    p_plus = 1/3
    p_minus = 1/3
    positions_new = positions.copy()
    for i in range(amount_places):
        if positions[i] == 0: continue
        for j in range(int(positions[i])):
            q = rd.uniform(0,1)
            if q <= p_minus:
                if i > 0:
                    positions_new[i-1] += 1
                    positions_new[i] -= 1
                else:
                    positions_new[amount_places-1] += 1
                    positions_new[i] -= 1
                n_minus += 1
            elif q > (1-p_plus):
                if i < (amount_places - 1):
                    positions_new[i+1] += 1
                    positions_new[i] -= 1
                else:
                    positions_new[0] += 1
                    positions_new[i] -= 1
                n_plus += 1
    J = (n_plus - n_minus)/amount_particles
    positions = positions_new.copy()
    return positions, J
```

```python
def ratchet_current(init_position, x_axis, alpha, T, Cycles, bk, N):
    # init_position: Initial particle distribution
    # x_axis: The part of the x-axis over which the particles are distributed. Same shape as init_position
    # T: Temporal period
    # N: Spacial period

    # First check if shapes match
    assert len(x_axis) == len(init_position)

    amount_particles = np.sum(init_position)
    amount_places = len(x_axis)
    x = x_axis

    params = (alpha, N)
    V = periodic(ratchet, -(1-alpha)*N, N, x, *params)
    positions = init_position

    J = []
    c = 0
    while c < Cycles:
        t = 0
        j = 0
        while t < 2*T:
            if t < T: #konstant
                out = const_step(positions, amount_places, amount_particles)
                positions = out[0]
                j += out[1]
            else: #sagtann
                out = periodic_step(positions, V, bk, amount_places, amount_particles)
                positions = out[0]
                j += out[1]
            t += 1
        c += 1
        J.append(j/(2*T))
    J = np.array(J)
    return positions, J
```

#### Fysisk tolkning av ikke-null strøm:
Vi får ikke-null strøm fordi systemet bytter periodisk mellom $V_2$ og $V_1$, der potensialet $V_1$ har en sagtannform. Når partiklene befinner seg i $V_2$ er de ikke bundet og vil spre seg ved tilfeldig vandring (diffusjon). I potensialet $V_1$ blir partiklene fanget nær bunnen av nærmeste sagtann-minimum.

Når partiklene deeksiteres til $V_1$ er det størst sjanse for at de havner på den slake delen av sagtanna, ettersom denne dekker størst andel av x-aksen. De fleste partiklene vil dermed bevege seg mot bunnpunktet til den slake delen av sagtanna. For $\alpha > 0.5$ vil venstre side av tanna ha slakest stigning og dermed gi netto transport mot venstre, slik at $n^{-}(t_i) > n^{+}(t_i)$ og fra (2) så vil $J(t_i)<0$, slik som vi observerer for $α=0.8$.

### Sammenligning mellom $J_{\mathrm{avg}}(0)$ og $J_{\mathrm{avg}}(n)$ for $n>0$:
Når den første syklusen $J_{\mathrm{avg}}(0)$ starter, så er alle partikler jevnt fordelt langs x-aksen i potensialet $V_2$, med seks partikler for hvert diskrete x-punkt. Når neste syklus starter vil partiklene ikke lenger ha en uniform fordeling, de vil være sentrert rundt bunnen av nærmeste sagtann-minimum. Dette fører til at $J_{\mathrm{avg}}(0)$ kan avvike fra de senere syklusene der partikkelfordelingen blir omorganisert og etterhvert vil oppnå en periodisk stasjonær fordeling. Som vil si at fordelingen ved starten av hver syklus vil være omtrent den samme.
$J_{\mathrm{avg}}(n)$ for $n>0$ blir da mer stabil (nær konstant), med små variasjoner på grunn av støy.

```python
def analytical_current(alpha, N, T):
    C = N/(4*T)
    a = ((alpha*N)/2) * (np.sqrt(3/T))
    b = (((1-alpha)*N)/2) * (np.sqrt(3/T))

    return C*(math.erfc(a) - math.erfc(b))
```

```python
def get_index_gap(b: float) -> int:
    """
    ε here is used as a small padding to effectivly get the same effect
    as >ceil(b). So instead of checking "if pos > ceil(b)", we get the
    index gap to place the proteins directly.

    Note:
        This is technically wrong and will not work correctly for numbers
        close to an index, e.g., 0.9999999. But we assume this will never
        be used.

    Params:
        b: Width of motor protein.
    """
    ε = 1e-6 # ε used to get greater than
    return math.ceil(b + ε)


def place_motor_proteins(b, start_position: int, amount: int, width: int) -> np.ndarray:
    """
    Place motor proteins on a line.

    Params:
        b: Width of a motor protein.
        start_position: Where to start placing motor proteins.
        amount: Amount of motor proteins.
        width: Width of np.ndarray.
    """
    line = np.zeros(width)
    index_gap = get_index_gap(b)
    half_index_gap = math.ceil(b/2)
    start_pos = start_position + half_index_gap

    # place motor protein with a gap of b between each protein
    for i in range(start_pos, width, index_gap + 1):
        if sum(line) == amount: break # when list has correct amount of particles
        line[i] = 1

    return line

def place_motor_proteins_uniformly(b, amount: int, x_axis: np.ndarray) -> np.ndarray:
    """
    Place motor proteins uniformly on a line.

    Params:
        b: Width of a motor protein.
    """
    width = len(x_axis)

    gap = width / amount
    init_positions = np.zeros(width)
    indices = np.arange(0,width,gap)
    for i in indices:
        init_positions[int(i)] = 1

    assert b < gap, "for trangt"

    return init_positions

def is_particle_to_the_left(line: np.ndarray, index: int, b: float) -> bool:
    """
    Check if the particle to the left is at a distance d > b from current particle.

    Params:
        line: Line with particles.
        index: Index of current particle.
        b: Width of a motor protein.
    """
    index_gap = get_index_gap(b)
    l = np.roll(line, +index_gap)
    if int(l[index]) == 1: return True
    return False

def is_particle_to_the_right(line: np.ndarray, index: int, b: float) -> bool:
    """
    Check if the particle to the right is at a distance d > b from current particle.

    Params:
        line: Line with particles.
        index: Index of current particle.
        b: Width of a motor protein.
    """
    index_gap = get_index_gap(b)
    l = np.roll(line, -index_gap)
    if int(l[index]) == 1: return True
    return False
```

```python
def ratchet_step_interaction(
        positions: np.ndarray,
        V: np.ndarray,
        b: float,
        bk: float,
    ) -> tuple[np.ndarray, np.float64]:
    """
    Simulate random steps given a ratchet potential.

    Params:
        positions: line with particles placed
        V: np.ndarray with the ratchet step potential
        b: width of motor protein
        bk: constant
    """

    # W. potential gradient
    n_plus = 0
    n_minus = 0
    positions_new = positions.copy()
    amount_places = len(positions_new)

    # get list of indices for positions there are particles
    particle_indices = [index for index, val in enumerate(positions_new) if val == 1]

    # scramble the list to get random order of the indices
    rd.shuffle(particle_indices)

    for i in particle_indices:

        if 0 < i < (amount_places-1): #checks if in the middle
            p_minus = (1 + np.e**(-bk*(V[i+1]-V[i-1])) + np.e**(-bk*(V[i]-V[i-1])))**(-1)
            p_plus = (1 + np.e**(-bk*(V[i-1]-V[i+1])) + np.e**(-bk*(V[i]-V[i+1])))**(-1)

        elif i == 0: #checks if in left boundary
            p_minus = (1 + np.e**(-bk*(V[i+1]-V[amount_places-1])) + np.e**(-bk*(V[i]-V[amount_places-1])))**(-1)
            p_plus = (1 + np.e**(-bk*(V[amount_places-1]-V[i+1])) + np.e**(-bk*(V[i]-V[i+1])))**(-1)

        elif i == (amount_places-1): #checks if in right boundary
            p_minus = (1 + np.e**(-bk*(V[0]-V[i-1])) + np.e**(-bk*(V[i]-V[i-1])))**(-1)
            p_plus = (1 + np.e**(-bk*(V[i-1]-V[0]))+np.e**(-bk*(V[i]-V[0])))**(-1)

        q = rd.uniform(0,1)
        if q <= p_minus and not is_particle_to_the_left(positions_new,i,b):
            at_left_boundary = i > 0
            if at_left_boundary:
                positions_new[i-1] += 1
                positions_new[i] -= 1
            else:
                positions_new[amount_places-1] += 1
                positions_new[i] -= 1
            n_minus += 1
        elif q > (1-p_plus) and not is_particle_to_the_right(positions_new,i,b):
            at_right_boundary = i < (amount_places - 1)
            if at_right_boundary:
                positions_new[i+1] += 1
                positions_new[i] -= 1
            else:
                positions_new[0] += 1
                positions_new[i] -= 1
            n_plus += 1
    J = (n_plus - n_minus)/np.sum(positions_new)
    positions = positions_new.copy()
    return positions, J


def const_step_interaction(
        positions: np.ndarray,
        b: float
        ) -> tuple[np.ndarray, np.float64]:
    """
    One step with constant potential $V(x) = k$

    Params:
        positions: list with amount of particles at each step.
        b: width of motor
    """
    n_minus = 0
    n_plus = 0
    p_plus = 1/3
    p_minus = 1/3
    positions_new = positions.copy()
    amount_places = len(positions_new)

    # get list of indices for positions there are particles
    particle_indices = [index for index, val in enumerate(positions_new) if val == 1]

    # scramble the list to get random order of the indices
    rd.shuffle(particle_indices)

    for i in particle_indices:
        there_is_no_particle = positions[i] == 0
        if there_is_no_particle: continue
        q = rd.uniform(0,1)
        if q <= p_minus and not is_particle_to_the_left(positions_new,i,b):
            if i > 0:
                positions_new[i-1] += 1
                positions_new[i] -= 1
            else:
                positions_new[amount_places-1] += 1
                positions_new[i] -= 1
            n_minus += 1
        elif q > (1-p_plus) and not is_particle_to_the_right(positions_new,i,b):
            if i < (amount_places - 1):
                positions_new[i+1] += 1
                positions_new[i] -= 1
            else:
                positions_new[0] += 1
                positions_new[i] -= 1
            n_plus += 1
    J = (n_plus - n_minus)/np.sum(positions_new)
    positions = positions_new.copy()
    return positions, J
```

```python
def flashing_ratchet_hardcore_repulsion(init_position, x_axis, alpha, T, b, Cycles, bk, N):
    assert len(x_axis) == len(init_position)

    amount_particles = np.sum(init_position)
    amount_places = len(x_axis)
    x = x_axis

    params = (alpha, N)
    V = periodic(ratchet, -(1-alpha)*N, N, x, *params)
    positions = init_position

    positions_2d = []
    J = []
    c = 0

    while c < Cycles:
        t = 0
        j = 0
        while t < 2*T:
            if t < T: #konstant
                out = const_step_interaction(positions, b)
                positions = out[0]
                j += out[1]
            else: #sagtann
                out = ratchet_step_interaction(positions, V, b, bk)
                positions = out[0]
                j += out[1]
            #if t == T:
            #    plt.scatter(x,positions*c,label=f"positions etter {c} iterasjoner")
            t += 1
            positions_2d.append(positions)
        c += 1
        #plt.scatter(x,positions*c,label=f"positions etter {c} iterasjoner")
        J.append(j/(2*T))
    J = np.mean(J)
    return J, positions_2d
```

```python
Cycles = 5
bk = 1000
N = 20
alpha = 0.2
T = 40
b = 2

params = (alpha, N)
x = np.arange(-2*N, 2*N, 1)
V = periodic(ratchet, -(1-alpha)*N, N, x, *params)
amount_particles = 10

init_position = place_motor_proteins_uniformly(b,amount_particles,x)

positions_2d = flashing_ratchet_hardcore_repulsion(init_position, x, alpha, T, b, Cycles, bk, N)[1]

for i in range(Cycles*2*T):
    plt.scatter(x,positions_2d[i]*(i+1),label=f"positions etter {i} cycles",color="lightseagreen",s=5)

plt.plot(x,V*100,color = "black")
plt.xlabel("$x$")
plt.ylabel("$t$",rotation="horizontal")
plt.title("Motorproteiners bevegelse over tid")
#plt.legend()
plt.show()
```
